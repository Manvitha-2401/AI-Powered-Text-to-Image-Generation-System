# Task 3 – Text Preprocessing using Hugging Face Transformers

## Objective

The objective of this task is to preprocess textual descriptions using Hugging Face Transformers by converting flower descriptions into numerical representations suitable for text-to-image generation.

The workflow includes:

- Loading flower descriptions
- Tokenization
- Attention Mask generation
- Text Embedding generation
- Preparing embeddings for the image generation model

## Note

The Oxford-102 Flowers dataset does not provide natural-language captions for each image. Therefore, representative flower descriptions were created to demonstrate the Hugging Face text preprocessing pipeline, including tokenization, attention mask generation, embedding creation, and preparation for text-to-image generation.

```python
from transformers import AutoTokenizer, AutoModel
import torch
```

```python
print("Libraries imported successfully!")
```

```python
# Sample flower descriptions

flower_descriptions = [
    "A red rose with green leaves",
    "A bright yellow sunflower under sunlight",
    "A white lily blooming in a garden",
    "A pink lotus floating on water",
    "A purple orchid with long petals"
]

print("Flower Descriptions:\n")

for desc in flower_descriptions:
    print("-", desc)
```

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

print("Tokenizer Loaded Successfully!")
```

```python
# Tokenize the flower descriptions

encoded_inputs = tokenizer(
    flower_descriptions,
    padding=True,
    truncation=True,
    return_tensors="pt"
)

print("Input IDs:\n")
print(encoded_inputs["input_ids"])

print("\nAttention Mask:\n")
print(encoded_inputs["attention_mask"])
```

```python
# Load the pre-trained DistilBERT model

model = AutoModel.from_pretrained("distilbert-base-uncased")

print("DistilBERT Model Loaded Successfully!")
```

```python
# Generate text embeddings

with torch.no_grad():
    outputs = model(**encoded_inputs)

embeddings = outputs.last_hidden_state

print("Embedding Shape:", embeddings.shape)
print("\nSample Embedding Vector:\n")
print(embeddings[0][0][:10])
```

## Integration of Text Embeddings into Image Generation Pipeline

The generated text embeddings are used as semantic representations of the textual descriptions. These embeddings are passed to the image generation model (Generator) to guide the image synthesis process.

Workflow:

Text Description
      ↓
Tokenizer
      ↓
Input IDs + Attention Mask
      ↓
DistilBERT
      ↓
Text Embeddings
      ↓
Generator (CGAN / Stable Diffusion)
      ↓
Generated Flower Image

```python
# Simulated integration of embeddings into the image generation model

generator_input = embeddings.mean(dim=1)

print("Generator Input Shape:", generator_input.shape)

print("\nThe generated text embeddings are now ready to be passed")
print("to the image generation model.")
```

## Conclusion

In this task, flower descriptions were successfully preprocessed using Hugging Face Transformers.

The following steps were completed:

- Loaded flower descriptions
- Tokenized text using DistilBERT tokenizer
- Generated Input IDs and Attention Masks
- Created contextual text embeddings
- Prepared embeddings for the image generation pipeline

These embeddings serve as the semantic input for the text-to-image generation model.
