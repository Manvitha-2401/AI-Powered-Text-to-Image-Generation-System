## Step 1: Import Required Libraries

Import the required Hugging Face libraries for tokenization and text encoding.

```python
from transformers import AutoTokenizer, AutoModel
import torch

print("Libraries imported successfully.")
```

## Step 2: Load Pre-trained Tokenizer and Model

We use DistilBERT from Hugging Face to tokenize text and generate embeddings.

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
model = AutoModel.from_pretrained("distilbert-base-uncased")

print("Tokenizer and model loaded successfully.")
```

## Step 3: Create Sample Flower Descriptions

The Oxford-102 Flowers dataset contains class labels but no text captions. Therefore, sample descriptions are created for demonstration purposes.

```python
flower_descriptions = [
    "A beautiful red flower.",
    "A bright yellow sunflower.",
    "A white daisy in the garden.",
    "A colorful tulip blooming in spring.",
    "A purple orchid with delicate petals."
]

for i, text in enumerate(flower_descriptions, start=1):
    print(f"{i}. {text}")
```

## Step 4: Tokenize Text Descriptions

The tokenizer converts each sentence into numerical representations called Input IDs and generates Attention Masks required by transformer models.

```python
encoded_data = tokenizer(
    flower_descriptions,
    padding=True,
    truncation=True,
    return_tensors="pt"
)

print("Tokenization completed successfully.")
```

## Step 5: Display Tokenized Output

```python
print("Input IDs\n")
print(encoded_data["input_ids"])

print("\nAttention Mask\n")
print(encoded_data["attention_mask"])
```

## Step 6: Generate Encoded Representations (Embeddings)

The DistilBERT model converts tokenized text into dense vector embeddings. These embeddings are used as input to text-to-image generation models.

```python
outputs = model(**encoded_data)

print("Embeddings generated successfully.")
```

## Step 7: Display Embedding Shape

```python
print("Embedding Shape:")
print(outputs.last_hidden_state.shape)
```
