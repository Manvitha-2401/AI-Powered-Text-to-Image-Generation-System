## Text Encoding using CLIP

The textual prompt is converted into contextual embeddings using the CLIP tokenizer and CLIP text encoder.

These embeddings provide semantic information that is later used by the Attention Generator to synthesize images corresponding to the textual description.

```python
tokenizer = CLIPTokenizer.from_pretrained(
    "openai/clip-vit-base-patch32"
)

text_encoder = CLIPTextModel.from_pretrained(
    "openai/clip-vit-base-patch32"
).to(device)

print("CLIP Loaded Successfully")
```

```python
class AttentionGenerator(nn.Module):
    def __init__(self, noise_dim=100, text_embedding_dim=512):
        super().__init__()

        self.fc = nn.Sequential(
            nn.Linear(noise_dim + text_embedding_dim, 1024),
            nn.ReLU(),

            nn.Linear(1024, 2048),
            nn.ReLU(),

            nn.Linear(2048, 3 * 64 * 64),
            nn.Tanh()
        )

    def forward(self, noise, attended_text):
        attended_text_flat = attended_text.mean(dim=1)

        x = torch.cat([noise, attended_text_flat], dim=1)

        x = self.fc(x)

        return x.view(-1, 3, 64, 64)
```

```python
from transformers import CLIPTokenizer, CLIPTextModel

tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-base-patch32")

text_encoder = CLIPTextModel.from_pretrained(
    "openai/clip-vit-base-patch32"
).to(device)

print("✅ CLIP Loaded Successfully")
```

```python
prompt = "A beautiful yellow sunflower"

print("Prompt:", prompt)
```

```python
inputs = tokenizer(
    prompt,
    return_tensors="pt",
    padding=True,
    truncation=True
)

inputs = {k: v.to(device) for k, v in inputs.items()}

print("Text Tokenization Complete")
```

```python
with torch.no_grad():

    embedding = text_encoder(**inputs).last_hidden_state

print("Embedding Shape:", embedding.shape)
```

```python
import torch.nn as nn

class CrossAttention(nn.Module):
    def __init__(self, image_dim=512, text_dim=512, num_heads=8):
        super().__init__()

        self.cross_attention = nn.MultiheadAttention(
            embed_dim=image_dim,
            num_heads=num_heads,
            batch_first=True
        )

    def forward(self, image_features, text_features):

        attended_features, attention_weights = self.cross_attention(
            query=image_features,
            key=text_features,
            value=text_features
        )

        return attended_features, attention_weights

print("✅ Cross-Attention Module Created Successfully")
```

```python
attention = CrossAttention().to(device)

print("✅ Attention Model Ready")
```

```python
image_features = torch.randn(1, 6, 512).to(device)

print("Image Features Shape:", image_features.shape)
```

```python
attended_text, attention_weights = attention(
    image_features,
    embedding
)

print("Attended Text Shape:", attended_text.shape)
```

```python
noise = torch.randn(1, 100).to(device)

print("Noise Shape:", noise.shape)
```

```python
with torch.no_grad():
    generated_image = generator(noise, attended_text)

print("Generated Image Shape:", generated_image.shape)
```

```python
import matplotlib.pyplot as plt

# Convert tensor to image
image = generated_image.squeeze(0).detach().cpu()

# Change shape from (3,64,64) → (64,64,3)
image = image.permute(1, 2, 0).numpy()

# Convert from [-1,1] to [0,1]
image = (image + 1) / 2
image = image.clip(0, 1)

plt.figure(figsize=(6,6))
plt.imshow(image)
plt.title(prompt)
plt.axis("off")
plt.show()
```

```python
prompts = [
    "A beautiful red rose",
    "A bright yellow sunflower",
    "A pink lotus flower",
    "A white daisy"
]

for p in prompts:

    inputs = tokenizer(
        p,
        return_tensors="pt",
        padding=True,
        truncation=True
    )

    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        embedding = text_encoder(**inputs).last_hidden_state

        image_features = torch.randn(1, 6, 512).to(device)

        attended_text, _ = attention(image_features, embedding)

        noise = torch.randn(1, 100).to(device)

        generated = generator(noise, attended_text)

    image = generated.squeeze().cpu().permute(1,2,0).numpy()
    image = (image + 1)/2
    image = image.clip(0,1)

    plt.figure(figsize=(4,4))
    plt.imshow(image)
    plt.title(p)
    plt.axis("off")
    plt.show()
```

```python
import matplotlib.pyplot as plt

plt.imsave(
    "/content/generated_flower.png",
    image
)

print("✅ Image Saved Successfully")
```
