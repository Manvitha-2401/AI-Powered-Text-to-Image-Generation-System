# Task 4 – Dataset Analysis

## Objective

The objective of this task is to perform an exploratory analysis of the Oxford-102 Flowers dataset before training the text-to-image generation model.

The analysis includes:

- Dataset overview
- Number of classes
- Number of training images
- Image dimension statistics
- Class distribution
- Width distribution
- Height distribution
- Sample flower images

The purpose of this analysis is to understand the dataset characteristics, identify variations in image sizes, and prepare the data for model training.

```python
from torchvision.datasets import Flowers102
import matplotlib.pyplot as plt
from collections import Counter
import random
```

## Step 1: Load the Dataset

The Oxford-102 Flowers dataset is used for image classification and contains 102 flower categories. In this analysis, the training split is loaded for exploration.

```python
dataset = Flowers102(
    root="data",
    split="train",
    download=True
)

print("Dataset Loaded Successfully!")
print("Total Training Images:", len(dataset))
```

## Step 2: Display Sample Flower Images

The following images are randomly selected from the training dataset to understand the diversity of flower categories.

```python
plt.figure(figsize=(10,10))

random_indices = random.sample(range(len(dataset)), 9)

for i, idx in enumerate(random_indices):
    image, label = dataset[idx]

    plt.subplot(3,3,i+1)
    plt.imshow(image)
    plt.title(f"Class {label}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

## Step 3: Dataset Statistics

This section analyzes the number of images, total classes, and class labels present in the Oxford-102 Flowers training dataset.

```python
labels = [label for _, label in dataset]

print("Total Training Images:", len(labels))
print("Unique Classes:", len(set(labels)))
print("Class IDs:", sorted(set(labels)))
```

### Observation

- The training dataset contains **1020 images**.
- There are **102 unique flower classes**.
- Each class contains approximately **10 training images**, making the dataset balanced for training machine learning models.

## Step 4: Class Distribution Analysis

The following graph shows the number of images available in each flower class.

```python
from collections import Counter
import matplotlib.pyplot as plt

class_counts = Counter(labels)

plt.figure(figsize=(14,5))
plt.bar(class_counts.keys(), class_counts.values())

plt.title("Number of Images per Flower Class")
plt.xlabel("Flower Class")
plt.ylabel("Number of Images")

plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.savefig("class_distribution.png")

plt.show()

print("Class distribution graph saved successfully!")
```

### Observation

The graph shows that all flower classes have nearly the same number of training images. This balanced distribution helps reduce bias during model training and improves learning performance.

## Step 5: Image Resolution Analysis

Image resolution is an important characteristic of a dataset because machine learning models usually require images to have a fixed input size. This section analyzes the width and height of all training images.

```python
widths = []
heights = []

for image, _ in dataset:
    width, height = image.size
    widths.append(width)
    heights.append(height)

print("Minimum Width :", min(widths))
print("Maximum Width :", max(widths))
print("Average Width :", sum(widths)/len(widths))

print()

print("Minimum Height:", min(heights))
print("Maximum Height:", max(heights))
print("Average Height:", sum(heights)/len(heights))
```

### Observation

The dataset contains images of varying resolutions.

- Minimum Width: **500 pixels**
- Maximum Width: **919 pixels**
- Average Width: **624.49 pixels**

- Minimum Height: **500 pixels**
- Maximum Height: **993 pixels**
- Average Height: **537.78 pixels**

Since the images have different resolutions, preprocessing such as resizing will be required before training deep learning models.

## Step 6: Image Resolution Distribution

The following histograms visualize the distribution of image widths and heights in the training dataset.

```python
# Width Histogram
plt.figure(figsize=(8,5))
plt.hist(widths, bins=20)
plt.title("Distribution of Image Widths")
plt.xlabel("Width (pixels)")
plt.ylabel("Number of Images")
plt.grid(alpha=0.3)
plt.savefig("width_distribution.png")
plt.show()

# Height Histogram
plt.figure(figsize=(8,5))
plt.hist(heights, bins=20)
plt.title("Distribution of Image Heights")
plt.xlabel("Height (pixels)")
plt.ylabel("Number of Images")
plt.grid(alpha=0.3)
plt.savefig("height_distribution.png")
plt.show()
```

### Observation

The histograms show that image dimensions vary across the dataset. Most images have widths around 600–650 pixels and heights around 500–550 pixels. This confirms that image resizing will be an important preprocessing step before training a text-to-image generation model.
